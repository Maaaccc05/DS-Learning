```python
import  seaborn as sns
import matplotlib.pyplot as plt
```

### There are 2 types 
1. Qualititive - Data which do not have much verification (For ex. Gender)
2. Quantitive - Data which have variations

```python
# Histplot/ distplot
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
# Distplot
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins = 20, kde = True) #actua value bin is 20 and distplot is also histplot, KDE(Kernel density estimator)
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins = 20, kde = True)
```

```python
# Jointplot to compare 2 datasets
sns.jointplot(x = 'total_bill', y = 'tip', data= df, kind = 'scatter')
```
